## **Theoritical Questions**

1. What is NumPy, and why is it widely used in Python?

- NumPy is a library for numerical computing. It’s fast and efficient for handling large arrays and matrices. It provides N-dimensional arrays, mathematical functions, broadcasting, and vectorized operations

2. How does broadcasting work in NumPy?

- Broadcasting lets NumPy perform operations on arrays of different shapes without explicit looping, by expanding smaller arrays to match the larger one’s shape.

3. What is a Pandas DataFrame?

- A Pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). It is a core data structure in the Pandas library for Python, widely used for data analysis and manipulation.

4.  Explain the use of the groupby() method in Pandas.

- The groupby function in Pandas is a tool that helps you organize data into groups based on certain criteria, like the values in a column.

5. Why is Seaborn preferred for statistical visualizations?

- Seaborn simplifies statistical data visualization in Python. Its integration with Pandas DataFrames and high-level plotting functions allow for quick and informative visualizations.

6. What are the differences between NumPy arrays and Python lists?

- NumPy arrays are fixed-type, homogeneous data structures that are optimized for numerical computations, while Python lists are dynamic, heterogeneous, and more flexible but slower. NumPy arrays support vectorized operations, enabling faster computations, whereas Python lists require explicit looping.

7. What is a heatmap, and when should it be used?

- A heatmap is a visual representation of data where values are depicted using colors, typically with warmer colors (like red or orange) indicating higher values and cooler colors (like blue or green) indicating lower values.

8.  What does the term “vectorized operation” mean in NumPy?

- It means performing operations on entire arrays without explicit Python loops, leveraging optimized C code for speed.

9. How does Matplotlib differ from Plotly?

- Matplotlib: Matplotlib is the go-to for highly customized, static visualizations for reports. Plotly: Interactive, web-based, supports zoom/hover, better for dashboards.

10. What is the significance of hierarchical indexing in Pandas?

- Allows multiple levels of indexing on rows/columns, making it easier to handle multi-dimensional data in a 2D structure.

11. What is the role of Seaborn’s pairplot() function?

- It creates a grid of scatterplots and histograms for pairwise relationships among multiple numerical variables.

12. What is the purpose of the describe() function in Pandas?

- Provides a quick statistical summary (count, mean, min, max, quartiles, std) of numerical or categorical data.

13. Why is handling missing data important in Pandas?

- Missing data can bias or break analysis. Pandas provides dropna(), fillna(), and interpolation to manage missing values.

14. What are the benefits of using Plotly for data visualization?

- Interactive visualizations, web integration, real-time dashboards, 3D plots, and ease of use for presentations.

15. How does NumPy handle multidimensional arrays?

- With the ndarray object, which can have any number of dimensions. It supports indexing, slicing, reshaping, and broadcasting in multiple dimensions.

16.  What is the role of Bokeh in data visualization?

- Bokeh specializes in interactive, web-friendly visualizations with streaming and real-time capabilities.

17. Explain the difference between apply() and map() in Pandas.

- map() works on a Series element-wise.

- apply() works on both Series and DataFrame, applying functions row-wise or column-wise.

18. What are some advanced features of NumPy?

- Broadcasting, vectorization, random number generation, linear algebra, FFT, masked arrays, memory mapping, and integration with C/C++ libraries.

19. How does Pandas simplify time series analysis?

- It provides built-in support for date/time indexing, resampling, frequency conversion, rolling windows, shifting, and period-based calculations.

20. What is the role of a pivot table in Pandas?

- Pivot tables summarize data by grouping and aggregating values.

21.  Why is NumPy’s array slicing faster than Python’s list slicing?

- Because NumPy uses contiguous memory blocks and slicing creates views (not copies), while Python lists store references and require more overhead.

22. What are some common use cases for Seaborn?

- Correlation analysis, distribution plots, categorical comparisons, regression visualization, heatmaps, and pairwise relationship exploration.

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import pandas as pd

#correlation matrix using Seaborn
data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
```

```python
correlation_matrix
```

```python
#visualize it with a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

13. Create a 3D scatter plot using Plotly.

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [10, 20, 30, 40, 50],
    'y': [10, 11, 12, 13, 14],
      'z': [55, 65, 75, 85, 95],
    'label': ['A', 'B', 'C', 'D', 'E']
})

fig = px.scatter_3d(df, x='x', y='y', z='z', color='label')
fig.show()
```
